## 1. QUES dataset

```python
import sklearn
import numpy as np
import pandas as pd
```

```python
data_train = pd.read_csv("D:/UROP/datasets/QUES.csv")
data_train.head()
```

```python
data_train.shape
```

```python
Y_train = pd.DataFrame(data = data_train.iloc[:, -1].values, columns = ['target'])
data_train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_train, Y_train, test_size=0.2, random_state=42)
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
X_train.mean()
```

```python
from sklearn import metrics
def print_error(X_test, y_test, model_name):
    prediction = model_name.predict(X_test)
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, prediction))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

### AdaBoost

```python
from sklearn.ensemble import AdaBoostRegressor

model = AdaBoostRegressor()

model.fit(X_train, y_train)
print_error(X_test,y_test, model)
```

```python
from sklearn.ensemble import AdaBoostRegressor

model = AdaBoostRegressor()

model.fit(X_train, y_train)
MRE = []
prediction = model.predict(X_test)
prediction = prediction.reshape(15,1)
MRE.append(abs(y_test-prediction)/y_test)

mre = str(MRE).split()
mre.remove("target")
mre_new = mre[2::2]
mre_new[len(mre_new)-1] = mre_new[len(mre_new)-1].strip("]")
for i in range(len(mre_new)):
    mre_new[i] = (float(mre_new[i]))
    
print(mre_new)

MMRE = sum(mre_new)/len(y_test)
print("MMRE is: ",MMRE)
count=0
for i in range(len(mre_new)):
    if(mre_new[i]<=0.3):
        count+=1
p=count/len(mre_new)
print("Pred(0.3) is", p)
```
